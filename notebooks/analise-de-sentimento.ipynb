```
from textblob import TextBlob
```

```
frase = "Python é ótimo para Machine Learning"
```

```
# instanciando um objeto chamado 'tb' (TextBlob) e inserir a frase dentro do objeto :
tb = TextBlob(frase)
```

```
# realizando a análise de sentimento da frase definida acima :
# importante : a ánalise de sentimento vai de -1 à +1, onde quanto mais próximo à -1 é uma menção negativa e +1 é uma menção positiva. "0" (zero) é neutro.
tb.sentiment.polarity
```

```
# um resultado igual à 'zero' indica neutralidade e, no caso acima em português, o TextBlob trabalha melhor com idioma inglês :
tb_en = tb.translate(to = 'en')
tb_en
```

```
tb_en.sentiment.polarity
# lembre-se : resultado próximo a '+1' indica que a menção é positiva !
```
